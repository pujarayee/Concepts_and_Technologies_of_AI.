Task - 3 - Outlier Detection:

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
import matplotlib.pyplot as plt

south_asia_df.loc['Score Z-score'] = (south_asia_df['score'] - south_asia_df['score'].mean()) / south_asia_df['score'].std()
south_asia_df.loc['GDP Z-score'] = (south_asia_df['Log GDP per capita'] - south_asia_df['Log GDP per capita'].mean()) / south_asia_df['Log GDP per capita'].std()

outliers_score = south_asia_df[abs(south_asia_df['Score Z-score']) > 2]
outliers_gdp = south_asia_df[abs(south_asia_df['GDP Z-score']) > 2]

outliers = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

print("Outlier Countries (based on Score and GDP per Capita):")
print(outliers[['Country name', 'Score Z-score', 'GDP Z-score']])

plt.figure(figsize=(10, 6))
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], label='Countries')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outliers')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('Outlier Detection in South Asia (Score vs. GDP)')
plt.legend()
plt.grid(True)
plt.show()
```

2. Define outliers using the 1.5 × IQR rule.

```python
# Calculate IQR for 'score'
Q1_score = south_asia_df['score'].quantile(0.25)
Q3_score = south_asia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Define outlier bounds for 'score'
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Identify outliers for 'score'
outliers_score_iqr = south_asia_df[(south_asia_df['score'] < lower_bound_score) | (south_asia_df['score'] > upper_bound_score)]

# Calculate IQR for 'Log GDP per capita'
Q1_gdp = south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Define outlier bounds for 'Log GDP per capita'
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers for 'Log GDP per capita'
outliers_gdp_iqr = south_asia_df[(south_asia_df['Log GDP per capita'] < lower_bound_gdp) | (south_asia_df['Log GDP per capita'] > upper_bound_gdp)]

# Combine outliers from both columns
outliers_iqr = pd.concat([outliers_score_iqr, outliers_gdp_iqr]).drop_duplicates()

(outliers_iqr[['Country name', 'score', 'Log GDP per capita']])
```

• Create scatter plots to visualize and interpret the relationships.

```python
import numpy as np
import matplotlib.pyplot as plt

# Ensure data is clean and numeric
def clean_data(df, x_col, y_col):
    df = df.dropna(subset=[x_col, y_col])  # Drop rows with NaN values in relevant columns
    df[x_col] = pd.to_numeric(df[x_col], errors='coerce')  # Ensure numeric data
    df[y_col] = pd.to_numeric(df[y_col], errors='coerce')
    return df.dropna(subset=[x_col, y_col])

# Plot scatter and regression
def plot_scatter_with_trendline(df, x_col, y_col, title, x_label, y_label):
    plt.figure(figsize=(8, 6))
    plt.scatter(df[x_col], df[y_col], label="Data Points")

    # Add labels and title
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)

    # Fit and plot regression line
    if len(df) >= 2:  # Ensure at least 2 points for regression
        z = np.polyfit(df[x_col], df[y_col], 1)
        p = np.poly1d(z)
        plt.plot(df[x_col], p(df[x_col]), "r--", label="Trend Line")
    else:
        print(f"Not enough data points for regression in {title}.")

    # Add grid and legend
    plt.grid(True)
    plt.legend()
    plt.show()

# Scatter plot for Freedom to Make Life Choices vs. Score (South Asia)
south_asia_df = clean_data(south_asia_df, 'Freedom to make life choices', 'score')
plot_scatter_with_trendline(south_asia_df, 'Freedom to make life choices', 'score',
                            'Freedom vs. Happiness Score (South Asia)',
                            'Freedom to Make Life Choices', 'Happiness Score')

# Scatter plot for Generosity vs. Score (South Asia)
south_asia_df = clean_data(south_asia_df, 'Generosity', 'score')
plot_scatter_with_trendline(south_asia_df, 'Generosity', 'score',
                            'Generosity vs. Happiness Score (South Asia)',
                            'Generosity', 'Happiness Score')

# Scatter plot for Freedom to Make Life Choices vs. Score (Middle East)
middle_east_df = clean_data(middle_east_df, 'Freedom to make life choices', 'score')
plot_scatter_with_trendline(middle_east_df, 'Freedom to make life choices', 'score',
                            'Freedom vs. Happiness Score (Middle East)',
                            'Freedom to Make Life Choices', 'Happiness Score')

# Scatter plot for Generosity vs. Score (Middle East)
middle_east_df = clean_data(middle_east_df, 'Generosity', 'score')
plot_scatter_with_trendline(middle_east_df, 'Generosity', 'score',
                            'Generosity vs. Happiness Score (Middle East)',
                            'Generosity', 'Happiness Score')
```

6. Outlier Detection:

• Identify outlier countries in both regions based on Score and GDP per Capita.

```python
import matplotlib.pyplot as plt

def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```

```python
# Identify outliers in South Asia based on Score
south_asia_outliers_score = identify_outliers(south_asia_df, 'score')
print("South Asia Outliers (Score):")
(south_asia_outliers_score[['Country name', 'score']])
```
